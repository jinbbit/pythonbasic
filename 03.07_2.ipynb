- 이번에는 콤마로 구분되지 않은 텍스트 파일에 대해서 처리해 보도록 하겠습니다. 
- 주피터 랩에서 매직 명령어인 ‘%%writefile 파일명’을 사용하여 파일을 저장할 수 
있습니다. 
- 아래의 코드를 작성하여 파일을 저장해봅시다.

```python
%%writefile datas/sample3.txt
c1        c2        c3      c4
0.179181 -1.538472 1.347553 0.43381
1.024209 0.087307 -1.281997 0.49265
0.417899 -2.002308 0.255245 -1.10515
```

- 이제 앞에서 만든 텍스트 파일을 불러오려면, 데이터를 구분하는 구분자(separator)가 
콤마(comma)가 아니면 sep 인수를 써서 구분자를 사용자가 지정해줘야 합니다. 
- 만약 길이가 정해지지 않은 공백이 구분자인 경우에는 ‘\s+’ 정규식(regular expression) 문자열을 사용하면 됩니다. //(띄어쓰기(공백) 하나 이상인걸 '정규식'으로 표현)

```python
pd.read_table("datas/sample3.txt", sep = '\s+')
```

- 이번에는 파일 안 내용을 살펴봤을 때 데이터뿐만 아니라 상단에 부가적인 텍스트가 
있는 경우를 살펴보려 합니다. 
- 아래의 코드를 작성하여 파일을 저장해보세요. 

```python
%%writefile datas/sample4.txt
파일 제목: sample4.txt
데이터 포맷의 설명:
c1, c2, c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

### < Pandas csv로부터 데이터 입력하기 >
- 데이터로 불러올 자료 파일 중에 건너 뛰어야 할 상단 행이 있으면 skiprows 인수를 
사용하면 됩니다. 
- 건너 뛸 줄을 리스트 안에 작성하면 됩니다. 리스트가 아닌 range(2)를 
활용할 수도 있습니다.

```python
pd.read_csv("datas/sample4.txt", sep = '\s+')
```

```python
# 리스트 안에 skiprows 인수 사용
pd.read_csv('datas/sample4.txt', skiprows = [0, 1])
```

```python
# 리스트가 아닌 range(2) 사용  어케쓰지??
pd.read_csv('datas/sample4.txt').range(2)
```

- 데이터로 불러올 자료 안 특정한 값을 NaN으로 취급하고 싶으면 na_values 인수에 
NaN 값으로 취급할 값을 넣습니다.

```python
df_read
```

```python
df_na_val = pd.read_csv("datas/sample1.csv", na_values = ["누락"])
df_na_val
```

- 다음은 웹사이트에 저장되어 있는 데이터 파일을 원격으로 읽는 명령 예제입니다.

```python
# 온라인의 csv 파일 가져오기 예제
country = pd.read_csv("https://raw.githubusercontent.com/datasets/country-codes/master/data/country-codes.csv")
country
```

```python
# 웅 안됑

import requests
import csv

# 데이터 다운로드
url = "https://raw.githubusercontent.com/datasets/country-codes/master/data/country-codes.csv"
response = requests.get(url)
content = response.content.decode('utf-8')

# csv 파일 읽어오기
csv_reader = csv.DictReader(content.splitlines(), delimiter=",")
countries = [row for row in csv_reader]

# 각 나라의 수도 출력하기
for country in countries:
    print(country['Name'], country['Capital'])
```
